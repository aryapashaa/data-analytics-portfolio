```
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/spotify.csv')
```

```
df
```

```
df.isna().sum()
```

```
df2 = df.copy()
```

```
df2.isna().sum()
```

```
df2 = df2.dropna()
```

```
df2.isna().sum()
```

```
df2
```

```
plt.figure(figsize=(15,7))

df2['Date'] = pd.to_datetime(df2['Date'])
df2.sort_values('Date', inplace=True)

plt.plot(df2['Date'],df2['Shape of You'], color = 'red', label = 'Shape of You')
plt.plot(df2['Date'],df2['Despacito'], color = 'orange', label = 'Despacito')
plt.plot(df2['Date'],df2['Something Just Like This'], color = 'yellow', label = 'Something Just Like This')
plt.plot(df2['Date'],df2['HUMBLE.'], color = 'green', label = 'HUMBLE.')
plt.plot(df2['Date'],df2['Unforgettable'], color = 'blue', label = 'Unforgettable')
plt.xlabel('Date')
plt.ylabel('Streams')
plt.title('Daily Global Streams of Popular Songs in 2017-2018')
plt.legend()
plt.show()
```
